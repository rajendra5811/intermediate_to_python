A list is a finite, ordered, mutable sequence of elements. Lists are one of the most fundamental, and most useful, collections that Python provides.

A list is delimited by square brackets, with commas separating elements

```python
list_A=['a',1,0.9,55,"good"]
list_A.append("bad")

print(list_A)
```

```python
x=[1,3,5,6,7,8]
y=[x,list_A]  #nested tuple
print(y[1][0])
print(y[0][2:5])
```

insert(),remove(),pop(),clear(),len() # methods

```python
x.insert(1,6) #(i,element)
print(x)
```

```python
x.pop(2)  #x=[1,3,5,6,7,8] 2 is index
print(x)
```

```python
6 in x
```

```python
x.remove(7) # 7 is element
print(x)
```

```python
x.reverse()
print(x)
```

```python
x.count(6)
```

Tuples are used to store collections of heterogenous immutable data.
A tuple is an immutable sequence of arbitrary data. 
A tuple is delimited by parentheses, and its elements are separated by commas. 

```python
tuple_a=(1,3,5,'good') # no assisgnment a[0]=10 !error
1 in tuple_a
print(tuple_a[3],tuple_a[3][1:2])
```

```python
v=([1,2,3,4],[6,7,8])
print(type(v))
print(len(v))
v[0].append(1)
print(v)
```

```python
employees = (
    (1, "CEO", "Madeline Bemelmans"),
    (2, "CTO", "Emily Elizabeth"),
    (3, "VP Ops", "François Borgot")
)
print(employees[1][0])
print(len(employees))
```

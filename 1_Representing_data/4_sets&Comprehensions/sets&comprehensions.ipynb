
A set is an unordered collection of distinct hashable elements.

```python
empty_set = set()
set_from_list = set([1, 2, 3, 2, 3, 4, 3, 1, 2]) # => {1, 2, 3, 4}
```

```python
# A basket of fruit names.
basket = {"apple", "orange", "apple", "pear", "banana"}

# How many unique fruit names are there?
len(basket)            # => 4

# Does the basket contain these elements?
"orange" in basket     # => True
"crabgrass" in basket  # => False

# Loop over the elements of the basket.
for fruit in basket:
    print(fruit )
```

```python
a = set("mississippi")  # {'i', 'm', 'p', 's'}

a.add('r')
a.remove('m')  # Raises a KeyError if 'm' is not present.
a.discard('x')  # Same as `remove`, except will not raise an error.

a.pop()  # => 's' (or 'i' or 'p')
a.clear()
len(a)  # => 0
```

```python
#Mathematical Set Operations
a = set("abracadabra")  # {'a', 'b', 'c', 'd', 'r'}
b = set("alacazam")     # {'a', 'm', 'c', 'l', 'z'}
#a[-1] error
# Set difference
print(a - b) # => {'b', 'd', 'r'}

# Union
print(a | b)  # => {'a', 'b', 'c', 'd', 'l', 'm', 'r', 'z'}

# Intersection
print(a & b) # => {'a', 'c'}

# Symmetric Difference
print(a ^ b)  # => {'b', 'd', 'l', 'm', 'r', 'z'}

# Subset
print(a <= b ) # => False
a.issubset
```

```python
"""Check whether a word is 'efficient' - if each letter can be drawn by a pencil without lifting."""
EFFICIENT_LETTERS = set('BCDGIJLMNOPSUVWZ')

def is_efficient(letters):
    return set(letters) <= EFFICIENT_LETTERS

def swap_keys_and_values(d):
    swapped = {}
    for key, value in d.items():
        swapped.setdefault(value, set()).add(key)
    return swapped
```

List Comprehensions
A list comprehension is a concise way to create and populate a list in Python in a declarative, rather than imperative, fashion.

```python
squares = []
for x in range(3):
    squares.append(x ** 2)
    print(squares)
```

```python
[x ** 2 for x in range(3)]
```

The general structure of a list comprehension is:

[f(xs) for xs in iter]

```python
[x ** 2 for x in range(3) if x % 2 == 0]
```

```python
sentence = ["This", "is", "a", "sentence"]

# Lowercased versions of words in a sentence.
[word.lower() for word in sentence]

# Words in a sentence of more than 8 letters long.
[word for word in sentence if len(word) > 8]

# 3-tuples of the first ten numbers and their squares and cubes.
[(x, x  **2, x**  3) for x in range(10)]

# Indices for a triangular matrix.
[(i,j) for i in range(5) for j in range(i)]
```

```python
d = {}
iterable = {1: "one", 2: "two", 3: "three"}

for vars in iterable:
    # vars is the key (e.g., 1)
    # iterable[vars] is the value (e.g., "one")
    
    key_func = iterable[vars]  # The new key becomes the old value ("one")
    value_func = vars          # The new value becomes the old key (1)
    
    d[key_func] = value_func

print(d)
# Output: {'one': 1, 'two': 2, 'three': 3}
```

```python
iterable = {1: "one", 2: "two", 3: "three"}

# This directly swaps the keys (k) and values (v)
d = {v: k for k, v in iterable.items()}

print(d)
# Output: {'one': 1, 'two': 2, 'three': 3}
```
